```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import gzip

# Path to your generated events
lhe_file = "/home/jovyan/mg5amcnlo/hh/Events/run_01/unweighted_events.lhe.gz"

def calculate_mhh(file_path):
    mhh_list = []
    with gzip.open(file_path, 'rt') as f:
        in_event = False
        for line in f:
            if "<event>" in line:
                in_event = True
                event_particles = []
                continue
            if "</event>" in line:
                in_event = False
                # Filter for Higgs bosons (PDG ID 25)
                higgs = [p for p in event_particles if p[0] == 25]
                if len(higgs) == 2:
                    # Sum energies and momenta: E, px, py, pz are indices 9, 6, 7, 8
                    e_tot = higgs[0][9] + higgs[1][9]
                    px_tot = higgs[0][6] + higgs[1][6]
                    py_tot = higgs[0][7] + higgs[1][7]
                    pz_tot = higgs[0][8] + higgs[1][8]
                    mhh = np.sqrt(e_tot**2 - (px_tot**2 + py_tot**2 + pz_tot**2))
                    mhh_list.append(mhh)
                continue
            
            if in_event and not line.startswith('<'):
                parts = line.split()
                if len(parts) == 13:
                    event_particles.append([int(parts[0])] + [float(x) for x in parts[1:]])
    return mhh_list

# Generate plot
masses = calculate_mhh(lhe_file)
plt.figure(figsize=(8,5))
plt.hist(masses, bins=50, color='skyblue', edgecolor='black')
plt.title("Di-Higgs Invariant Mass $m_{hh}$")
plt.xlabel("Mass [GeV]")
plt.ylabel("Events")
plt.grid(alpha=0.3)
plt.show()
```
